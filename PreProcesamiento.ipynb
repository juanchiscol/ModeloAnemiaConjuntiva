#### Redimensionamiento (relleno)

```python
import math
from skimage.io import imsave, imread

for images in Data_files:
    img=np.array(cv2.imread(os.path.join(dir_path, images)))
    #print(img.shape)
    padding = ((math.floor((max-img.shape[0])/2), math.ceil((max-img.shape[0])/2)), (math.floor((max-img.shape[1])/2), math.ceil((max-img.shape[1])/2)),(0,0))
    y = np.pad(img, padding, mode='constant',constant_values=0)
    #print(y.shape)
    padded_image = y[:, :, [2, 1, 0]]
    imsave(r"C:\Users\HP\Desktop\ModeloAnemiaConjuntiva\Datapreprocesada" + "/" + images, padded_image)
```

```python
import os
import numpy as np
import cv2

# Define the directory path to the dataset
dir_path = r"C:\Users\HP\Desktop\ModeloAnemiaConjuntiva\Dataset"
dataset = []
files_folder = []

# Define the target size
target_size = (160, 160)  # Set the desired size (width, height)

# Load images from the directory
for file_path in os.listdir(dir_path):
    if os.path.isfile(os.path.join(dir_path, file_path)):
        img = cv2.imread(os.path.join(dir_path, file_path))
        
        if img is not None:  # Check if the image was loaded successfully
            img_resized = cv2.resize(img, target_size)  # Resize the image
            dataset.append(img_resized)
            files_folder.append(file_path)

# Convert the dataset list to a NumPy array
dataset = np.array(dataset)
print("Dataset shape:", dataset.shape)

# Initialize labels
labels = []

# Assign labels based on file names
for files in files_folder:
    if files[0] == 'N':
        labels.append(0)  # Label for 'N'
    elif files[0] == 'A':
        labels.append(1)  # Label for 'A'

# Convert labels list to a NumPy array
labels = np.array(labels)
print("Labels shape:", labels.shape)
```

```python
import pickle

# Guardar el dataset y las etiquetas usando rutas en formato raw string
pickle.dump(dataset, open(r'C:\Users\HP\Desktop\ModeloAnemiaConjuntiva\x_final_conjunctiva.pkl', 'wb'))
pickle.dump(labels, open(r'C:\Users\HP\Desktop\ModeloAnemiaConjuntiva\y_final_conjunctiva.pkl', 'wb'))
```
